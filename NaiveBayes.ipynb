```python
import pandas as pd
import numpy as np
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score

# Load the train and test datasets
train_df = pd.read_csv('SalaryData_Train.csv')
test_df = pd.read_csv('SalaryData_Test.csv')

```

```python
# Preprocess the data
from sklearn.preprocessing import LabelEncoder

# Convert categorical variables to numerical variables
label_encoder = LabelEncoder()
train_df['workclass'] = label_encoder.fit_transform(train_df['workclass'])
train_df['education'] = label_encoder.fit_transform(train_df['education'])
train_df['maritalstatus'] = label_encoder.fit_transform(train_df['maritalstatus'])
train_df['occupation'] = label_encoder.fit_transform(train_df['occupation'])
train_df['relationship'] = label_encoder.fit_transform(train_df['relationship'])
train_df['race'] = label_encoder.fit_transform(train_df['race'])
train_df['sex'] = label_encoder.fit_transform(train_df['sex'])
train_df['native'] = label_encoder.fit_transform(train_df['native'])
```

```python

test_df['workclass'] = label_encoder.fit_transform(test_df['workclass'])
test_df['education'] = label_encoder.fit_transform(test_df['education'])
test_df['maritalstatus'] = label_encoder.fit_transform(test_df['maritalstatus'])
test_df['occupation'] = label_encoder.fit_transform(test_df['occupation'])
test_df['relationship'] = label_encoder.fit_transform(test_df['relationship'])
test_df['race'] = label_encoder.fit_transform(test_df['race'])
test_df['sex'] = label_encoder.fit_transform(test_df['sex'])
test_df['native'] = label_encoder.fit_transform(test_df['native'])

```

```python
# Split the data into features and target variables
X_train = train_df.drop('Salary', axis=1)
y_train = train_df['Salary']
X_test = test_df.drop('Salary', axis=1)
y_test = test_df['Salary']
```

```python
# Build and train the model
gnb = GaussianNB()
gnb.fit(X_train, y_train)
```

```python
# Make predictions on the test data
y_pred = gnb.predict(X_test)

# Evaluate the accuracy of the model
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy}")
```
